# This notebook calculates power required to provide sound to a room
The total power required combines humidification power with sound amplification power

First, I use standard models for attenuation of sound in air at different humidities

Then, I calculate how much power is required to keep a room at some humidity above ambient humidity

The choice of humidity is a tradeoff to minimize this combined power 

```python
import numpy as np
from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
pr = 101.325 # kPa
pa = pr #kPA, ambient atmospheric pressure
To1 = 273.16 #K
To = 273.15 #K, reference temp
T = To+20 #K, = 20 deg C

def calculateAttenuationCoef(hr,f,pa=pr,T=To+20):
    psat = pr * 10** ( -6.8346 * (To1 / T)**1.261 + 4.6151 )
    h = hr * psat / pa

    frN = (pa / pr) * (T / To)**(-0.5) * (9 + 280 * h * np.exp (-4.170 * ((T / To)**(-1/3)-1)))
    frO = (pa / pr) * (24.0 + 4.04E4 * h * ((0.02 + h) / (0.391 + h)))
    z = 0.1068*np.exp(-3352/T)*((frN + f**2 / frN)**-1)
    y = (T / To)**(-5.0/2) * (0.01275*np.exp(-2239.1 / T) * (frO + f**2/frO)**-1 + z)
    a = 8.686 * f**2 * ((1.84E-11 * (pa/pr)**-1 * (T/To)**0.5 + y)) #attenuation dB/m
    return a

def calculatePAir(a,s,Pi):
    a_s = a*s
    x =  1/(10*np.log10((np.exp(1)))*2) #for converting distance to log distance?
    #attenuation experession
    Pt = Pi * np.exp(-x*a_s)
    return Pt

def dBlossAir(a,s):
    a_s = a*s
    x =  1/(10*np.log10((np.exp(1)))*2) #for converting distance to log distance?
    #attenuation experession
    DeltaLt = 20 * np.log10(1/np.exp(-x*a_s)) #attenuation due to atm absorbtion, dB
    return DeltaLt


# db loss over distance:
def dBlossDist(r2, r1=1):
    return 20* np.log10(r2/r1)

def dBToPowerRatio(dB):
    return 10**(dB/10)

def powerRatioTodB(powerRatio):
    return 10*np.log10(powerRatio)

def dBToSPLRatio(splRatio):
    return 10**(splRatio/20)

def splRatioTodB(powerRatio):
    return 20 *np.log10(powerRatio)
```

```python
#power consumption of the speaker system:
def wallPowerRequired(distance,dbWanted,sensitivity,ampEfficiency,hr,f):
    #attenuation: 
    a = calculateAttenuationCoef(hr,f)
    airLoss = dBlossAir(a,distance)
    
    #speaker power required
    speaker1mLevel = dbWanted + dBlossDist(distance) + airLoss 
    dbAboveSensitivity = speaker1mLevel - sensitivity
    speakerPower = dBToPowerRatio(dbAboveSensitivity)
    #wall power required
    wallPower = speakerPower/ampEfficiency
    return wallPower

#given power to system, what is the volume in the room?
def measuredSoundLevel(distance,sensitivity,ampEfficiency,hr,f,wallPower):
    speakerPower = wallPower * ampEfficiency
    #spl at 1 m for power provided:
    splAt1m = sensitivity + powerRatioTodB(speakerPower)
    
    #attenuation: 
    a = calculateAttenuationCoef(hr,f)
    airLoss = dBlossAir(a,distance)
    
    splAtDistance = splAt1m - dBlossDist(distance) - airLoss
    return splAtDistance

#given power to system, what is the volume in the room?
def measuredSoundLevelAvg(distance,sensitivity,ampEfficiency,hr,fRange,wallPower):
    
    frequencies = np.logspace(np.log10(fRange[0]),np.log10(fRange[1]),base=10,num=100)
    spl = []
    for f in frequencies:
        spl.append(measuredSoundLevel(distance,sensitivity,ampEfficiency,hr,f,wallPower))
    return np.mean(spl)

def soundLevelPenalty(splWanted,splMeasured):
    #minimum noticeable difference is about 3dB. Penalty should roughly compare to energy use
    penaltySize = .015
    delta = abs(splWanted-splMeasured)
    penalty = 1 + penaltySize * delta**2
    return penalty
```

```python
#we can assume our amp is an old amp that is inefficient:
ampEfficiency = 0.1 #10% efficient old amp

#distnace from speaker:
distance = 40 #m
dbWanted = 85 #dB SPL
sensitivity = 100 #dB SPL at 1 Wm, for this speaker
humidity = 50 # % relative humidity

h = np.linspace(0,100,100)
f = np.arange(500,3000,500)
for freq in f:
    power = []
    for humidity in h:
        power.append(wallPowerRequired(distance,dbWanted,sensitivity,ampEfficiency,humidity,freq))
    plt.plot(h,power,label='%s Hz' % freq)
plt.ylabel("P (W)")
plt.xlabel("Relative Humidity (%)")
plt.legend()
t = plt.title("Power required for 85dB at 40 m")
```

# Calculate Power To Humidifier

```python
#size of the room
l_room = 50 #m
w_room = 30 #m
h_room = 20 #m


def powerToHumidify(desiredH,outsideH,l_room,w_room,h_room):
    roomVolume = l_room * w_room * h_room # m^3
    #rate of fresh air from outside entering
    # cfm standard from  https://energycodeace.com/site/custom/public/reference-ace-2013/index.html#!Documents/43ventilationrequirements.htm 
    cfm = .15 #.15 Cubic feet per minute, per square ft floor space,
    airInPerHr = cfm * (l_room*h_room/.304**2) * .304**3 * 60 #m^3/per hour of outside air entering
    airWeightIn =  1.19 * airInPerHr #air weighs 1.19 kg/m3 at 20 C
#     turnover = airInPerHr / roomVolume

    #how much water do we need to add? 
    #at 20 C, kg H20 per kg air is: 14.62e-3 kg/kg * H (% relative humidity)
    water_toAdd = (desiredH-outsideH)/100 * 14.62E-3 * airWeightIn #kg per hr, equal to L/hr

    capacityHumidifier = 1.5 * 3.79# GPH * L/G gives Liters per Hr
    maxPowerConsumption_Humidifier = 200 #W, for TS100 in-duct model https://www.smartfog.com/product/ts100
    humidifierOperatingRate = water_toAdd / capacityHumidifier
    powerToHumidify = humidifierOperatingRate * maxPowerConsumption_Humidifier #we assume linear power with operating rate (% of total capacity)
    return powerToHumidify #Watts

#relative humidity ambient and inside (desired)
desiredH = 20
outsideH = 10
powerToHumidify(desiredH, outsideH, l_room, w_room, h_room)
```

# Calculate Total Power Required
Sum of Humidification Power and Sound Amplification power required at that humidity 

```python
import matplotlib.pylab as pl

distance = 40 #m
dbWanted = 85 
sensitivity = 100 #dB at Watt meter
ampEfficiency = 0.1
outsideH = 10
insideH = 20

allH = np.linspace(outsideH, outsideH+30,4)
frequencies = np.linspace(1000,8000,100)
colors = pl.cm.jet(np.linspace(0,1,len(allH)*1.5))

avgPower = []
for i in range(len(allH)):
    insideH = allH[i]
    totalP = []
    for freq in frequencies:
        humidifierPower = powerToHumidify(insideH, outsideH, l_room, w_room, h_room)
        speakerPower = wallPowerRequired(distance,dbWanted,sensitivity,ampEfficiency,insideH,freq)
        totalP.append(humidifierPower + speakerPower)
    plt.plot(frequencies, totalP,label='H=%s' % insideH, c=colors[i])
    avgPower.append(np.mean(totalP))
plt.legend()
plt.xlabel("frequency")
plt.ylabel("total power required")
plt.show()
```

# Cost Function
Sum of Humidification Power and Sound Amplification power required at that humidity, plus a penalty for differing from desired SPL

```python
def objective(powerToSpeaker,humidity):
    distance = 50 #m
    spl_desired = 90
    sensitivity = 100 #dB at Watt meter
    ampEfficiency = 0.1
    outsideH = 20
    l_room = 50
    w_room = 30
    h_room = 20
    fRange = [2000,8000]
    
    powerToHumidifier = powerToHumidify(humidity,outsideH,l_room,w_room,h_room)
#     print(powerToHumidifier)
    totalPower = powerToSpeaker + powerToHumidifier
    
    spl = measuredSoundLevelAvg(distance,sensitivity,ampEfficiency,humidity,fRange,powerToSpeaker)
    penalty = soundLevelPenalty(spl, spl_desired)
    score = totalPower*penalty + 1000*(penalty-1)
    return score
```

```python
objective(100,30)
```

```python
for h in np.linspace(20,35,4):
    s = []
    power = np.linspace(500,1500,100)
    for p in power:
        s.append(objective(p,h))
    plt.plot(power,s, label='%s H' % h)
plt.legend()
```
